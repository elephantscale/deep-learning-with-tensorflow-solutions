# Setup your local laptop for the labs

This guide will walk you through how to setup a local setup to run the labs.

There are two ways of running labs
- Option A : Easiest if it works : Use training docker image
- Option B : Takes some work : Native setup

## Operating System
Setup is easier if you have either **MacOS or Linux**.   
If you are on Windows, try
- run the docker image if your machine supports docker
- Download and run ubuntu linux virtual machine using VirtualBox

---

## Option A : Using Docker Image

### Software Needed
- Docker
- that is it!

### Step A1 : Install Docker
Download Docker [from here](https://www.docker.com/) and install

### Step A2 : Test your Docker is working
- open a terminal
- enter command    
`docker images`  
see if it works
- Try this  
`docker run hello-world`  
This will download the docker image and run it.  
If this works, docker is working on your system

### Step A3 : Download training docker image
- Create an account at [Dockerhub](https://hub.docker.com/)
- Give the instructor your docker-hub-id so we can add it to the authorized user group
- Try the following

```bash 
$  docker login
# enter your credentials here

$  docker pull elephantscale/es-training
# this will take some time download the training image

$  docker images
# you should see the training image
```

### Step A4 : Download dataset
- You can access the data repository [here](https://s3.amazonaws.com/elephantscale-public/data/datasets.zip) (click link to download.  warning large file size ~300Meg + )  
To download this from command line  
`$    wget 'https://s3.amazonaws.com/elephantscale-public/data/datasets.zip'`
- Unzip the dataset in $HOME/datasets  
Also create a symlink like this  
`$   sudo ln -s ~/datasets /data `

### Step A5 : Download the labs / solutions
- instructor will provide the labs bundle
- unzip the lab (anywhere)

### Step A6 : Run the docker
- in the lab dir, you will see a `run-docker.sh`  
This is the script we will use to launch the docker image.
- It will do the following directory mappings
  - current lab directory on host machine --> `$HOME/dev` directory in docker
  - `$HOME/datasets` directory on host machine --> `/data` directory in docker

```bash
$   cd deep-learning-labs  # or where ever you unzipped the labs
$   ./run-docker.sh    elephantscale/es-training
```

### A7 : Access Jupyter Page
- default password is `bingobob123`
- In Jupyter, navigate to `dev` directory to see the files



### A8: Open and run `testing123.ipynb` file 
This file is under `0-testing` directory.   
This file will check your setup.  
If there are no errors here, then you are good to go

**You are up and running**  
**Happy learning!**

## Option B : Native Setup
This works best on **Mac or Linux**

### Software Needed
- Anaconda Python version 3.x


### B1: Install Anaconda Python
Download and install Anaconda Python version 3.x from [here](https://www.anaconda.com/download/).

### B2 : Setup a Tensorflow v2 environment
```bash
$ conda env list

# create a new env
$ conda create -y -n tf2 \
    ipykernel \
    jupyter \
    jupyterlab \
    matplotlib==3.1.0\
    numpy \
    opencv \
    pandas \
    pillow \
    pydot \
    scikit-learn\
    scipy \
    seaborn \
    unidecode

$ conda activate tf2

$ pip install --upgrade pip

$ pip install tensorflow==2.0.0-rc0 tensorboard  keras

$ ipython kernel install --user --name=tf2 
```

### B3: Download data files
- You can access the data repository [here](https://s3.amazonaws.com/elephantscale-public/data/datasets.zip) (click link to download.  warning large file size ~300Meg + )  
To download this from command line
```bash 
$    wget 'https://s3.amazonaws.com/elephantscale-public/data/datasets.zip'
```
- Unzip the dataset in `$HOME/datasets`  
Also create a symlink like this  
```bash
$   sudo ln -s ~/datasets /data
```

### B4 : Download the labs / solutions
- instructor will provide the labs bundle
- unzip the lab (anywhere)

### B5: Run  `run-jupyter.sh`
This file is located in the labs directory. 
```bash
./run-jupyter.sh
```

This will print out a URL, go to that URL.


### B6: Open and run `testing123.ipynb` file 
This file is under `0-testing` directory.   
This file will check your setup.  
If there are no errors here, then you are good to go

### B7 : And practice!
